## Date Range

```python
import pandas as pd
```

```python
df = pd.read_csv("apple_stock_price_nodate.csv")
df.head()
```

```python
#rng = pd.date_range(start="2018-07-23",end="2019-07-23",freq='b')
rng = pd.date_range(start="2018-07-23",periods=df.shape[0],freq='b')
rng
```

```python
df.set_index(rng, inplace=True)
df.head()
```

## Finding missing dates from datetimeindex

```python
daily_index = pd.date_range(start="7/1/2019",end="8/6/2019",freq='D')
daily_index
```

```python
daily_index.difference(df.index)
```

## asfreq

```python
df.index
```

```python
df.asfreq('D',method='pad')
```

```python
# Null value for weekends
df.asfreq(freq="D").head(10)
```

```python
# Fill the Nan value with the price of last friday
df.asfreq('D',method='pad').head(10)
```

```python
# Set frenq as Week
df.asfreq('W',method='pad').head(10)
```

```python
# Set frenq as Hour
df.asfreq('H',method='pad').head(24)
```
